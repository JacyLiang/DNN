```python
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
#載入資料
df = pd.read_csv('./output3.csv', encoding='utf-8')
df
```

```python
#取標籤
label = df["class"]
label
```

```python
#取出特徵
data = df.iloc[:,1:]
data
```

```python
#資料標準化
scaler = MinMaxScaler(feature_range=(0,1))
data = scaler.fit_transform(data)
data
```

```python
#分割資料
#把資料轉numpy
reshaped_data = np.array(data)
reshaped_label = np.array(label)
#將資料分為8:2

train_feature, test_feature, train_label, test_label = train_test_split(reshaped_data, reshaped_label, test_size=0.2, random_state = 42)

# split_boundary = int(reshaped_data.shape[0]*0.8)
# train_feature = reshaped_data[: split_boundary]
# test_feature = reshaped_data[split_boundary:]
# train_label = reshaped_label[: split_boundary]
# test_label = reshaped_label[split_boundary:]
print("測試資料:", train_feature)
print("訓練資料:", test_feature)
print("測試資料數量:", len(train_feature))
print("訓練資料數量:", len(test_feature))
print("測試資料維度:", train_feature.shape)
print("訓練資料維度:", test_feature.shape)
```

```python
#將訓練資料打散
```

```python
train_label = np_utils.to_categorical(train_label)
test_label= np_utils.to_categorical(test_label)
print("訓練標籤:",train_label.shape)
print("測試標籤:",test_label.shape)
```

```python
#更改資料維度
train_feature = np.reshape(train_feature, (train_feature.shape[0], train_feature.shape[1], 1))
test_feature = np.reshape(test_feature, (test_feature.shape[0], test_feature.shape[1], 1))
print("測試資料維度:", train_feature.shape)
print("訓練資料維度:", test_feature.shape)
```

```python
#訓練模型
model = Sequential()
n = 0.2
model.add(Dense(units=32, input_dim=60, activation='relu'))    #輸入60個特徵
model.add(Dropout(n))
model.add(Dense(units=64, activation='relu'))
model.add(Dropout(n))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(n))
model.add(Dense(units=256, activation='relu'))
model.add(Dropout(n))
model.add(Dense(units=5, activation='softmax'))   #輸出為5個分類
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
fitmodel = model.fit(x=train_feature, y=train_label, validation_split=0.2, epochs=40, batch_size=20, verbose=2)
score = model.evaluate(x=test_feature, y=test_label, verbose=2)
print('\n測試資料準確率:', score[1])
```
